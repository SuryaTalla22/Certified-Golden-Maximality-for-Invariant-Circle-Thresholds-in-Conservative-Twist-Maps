# Upper Ladder Refinement and Denominator-Aware Consistency

This notebook exercises the new upper-ladder refinement layer: cluster formation, denominator-aware consistency checks, and the refined upper-side object used by the two-sided irrational corridor.

```python
from pathlib import Path
import sys, math
ROOT = Path.cwd().resolve()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
print('ROOT =', ROOT)
```

```python
from kam_theorem_suite.obstruction_atlas import ApproximantWindowSpec
from kam_theorem_suite.proof_driver import (
    build_refined_rational_upper_ladder_report,
    build_two_sided_irrational_threshold_atlas_report,
)
from kam_theorem_suite.standard_map import HarmonicFamily

GOLDEN = (math.sqrt(5.0) - 1.0) / 2.0
fam = HarmonicFamily()
specs = [
    ApproximantWindowSpec(3, 5, 0.9738, 0.9742, 1.04, 1.06, label='gold-3/5'),
    ApproximantWindowSpec(5, 8, 0.96, 0.98, 1.04, 1.06, label='gold-5/8'),
    ApproximantWindowSpec(8, 13, 0.9710, 0.9725, 1.04, 1.06, label='gold-8/13'),
]
```

```python
rep = build_refined_rational_upper_ladder_report(
    rho=GOLDEN,
    specs=specs,
    family=fam,
    initial_subdivisions=2,
    max_depth=2,
)
{
    'status': rep['status'],
    'selected_cluster_member_labels': rep['refined_upper_ladder']['selected_cluster_member_labels'],
    'best_interval': (
        rep['refined_upper_ladder']['best_refined_crossing_lower_bound'],
        rep['refined_upper_ladder']['best_refined_crossing_upper_bound'],
    ),
    'improved_width': rep['refined_upper_ladder']['refinement_improved_width'],
}
```

```python
atlas = build_two_sided_irrational_threshold_atlas_report(
    rho=GOLDEN,
    K_values=[0.5, 0.6, 0.7],
    specs=specs,
    family=fam,
    N_values=(32, 48, 64),
    quality_floor='weak',
    initial_subdivisions=2,
    max_depth=2,
    refine_upper_ladder=True,
)
{
    'status': atlas['relation']['status'],
    'upper_object_source': atlas['relation']['upper_object_source'],
    'upper_object': (atlas['relation']['upper_object_lo'], atlas['relation']['upper_object_hi']),
    'has_refined_cluster': atlas['relation']['has_refined_cluster'],
}
```
