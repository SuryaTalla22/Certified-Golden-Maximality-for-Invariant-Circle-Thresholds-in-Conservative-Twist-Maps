# Automatic Class Campaigns

Generate ladders and atlases automatically from arithmetic classes.

```python
from pathlib import Path
import sys

PROJECT_ROOT = None
for cand in [Path.cwd(), Path.cwd().parent, *Path.cwd().parents]:
    if (cand / "kam_theorem_suite").exists():
        PROJECT_ROOT = cand
        break
if PROJECT_ROOT is None:
    raise RuntimeError("Could not locate project root containing kam_theorem_suite")
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
print("PROJECT_ROOT =", PROJECT_ROOT)
```

```python
from kam_theorem_suite import ArithmeticClassSpec
from kam_theorem_suite.proof_driver import build_class_ladder_report_driver, build_class_atlas_campaign_report

silver = ArithmeticClassSpec(preperiod=(), period=(2,), label="silver")
ladder = build_class_ladder_report_driver(silver, max_q=30, keep_last=3, q_min=2)
ladder_summary = {
    "class_label": ladder["class_label"],
    "eta_interval": [ladder["eta_lo"], ladder["eta_hi"]],
    "approximant_labels": [a["label"] for a in ladder["approximants"]],
}
ladder_summary
```

```python
campaign = build_class_atlas_campaign_report(
    silver,
    reference_lower_bound=0.97122,
    reference_crossing_center=0.9713,
    max_q=20,
    keep_last=2,
    q_min=2,
    auto_localize_crossing=True,
)
{
    "class_label": campaign["class_label"],
    "atlas_status": campaign["atlas"]["atlas_status"],
    "pruning_status": campaign["pruning_against_reference"]["status"],
}
```
