# Golden supercritical continuation

This notebook exercises the local continuation audit for the golden upper-side obstruction object and the resulting two-sided bridge.

```python
from pathlib import Path
import sys, json

ROOT = Path.cwd()
for candidate in [ROOT, ROOT.parent]:
    if (candidate / 'kam_theorem_suite').exists():
        ROOT = candidate
        break
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
print(ROOT)
```

```python
from kam_theorem_suite import (
    build_golden_supercritical_continuation_report,
    build_golden_two_sided_continuation_bridge_report,
)

continuation = build_golden_supercritical_continuation_report(
    n_terms=5,
    keep_last=2,
    min_q=5,
    crossing_center_offsets=(-4.0e-4, 0.0, 4.0e-4),
    initial_subdivisions=1,
    max_depth=0,
    refine_upper_ladder=False,
)
print(json.dumps({
    'status': continuation['theorem_status'],
    'usable_upper_step_count': continuation['usable_upper_step_count'],
    'stable_upper_support_size': continuation['stable_upper_support_size'],
}, indent=2))
```

```python
bridge = build_golden_two_sided_continuation_bridge_report(
    K_values=[0.3],
    N_values=(24,),
    use_multiresolution=False,
    n_terms=5,
    keep_last=2,
    min_q=5,
    crossing_center_offsets=(-4.0e-4, 0.0, 4.0e-4),
    initial_subdivisions=1,
    max_depth=0,
    refine_upper_ladder=False,
)
print(json.dumps({
    'status': bridge['theorem_status'],
    'relation': bridge['relation'],
}, indent=2))
```
