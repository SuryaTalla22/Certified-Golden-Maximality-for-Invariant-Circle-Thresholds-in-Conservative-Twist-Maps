# 17. Class-level Challenger Exhaustion Screen

This notebook aggregates the per-class asymptotic upper-ladder audits into a single exhaustion-facing screen across several challenger classes.

```python
from pathlib import Path
import sys

ROOT = Path.cwd()
if not (ROOT / 'kam_theorem_suite').exists():
    for cand in [ROOT, *ROOT.parents]:
        if (cand / 'kam_theorem_suite').exists():
            ROOT = cand
            break
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
print(ROOT)
```

```python
from kam_theorem_suite import ArithmeticClassSpec, build_class_exhaustion_screen_report

specs = [
    ArithmeticClassSpec(preperiod=(), period=(2,), label='silver'),
    ArithmeticClassSpec(preperiod=(), period=(3,), label='bronze'),
    ArithmeticClassSpec(preperiod=(1,), period=(2,), label='noble-like'),
]
report = build_class_exhaustion_screen_report(
    specs,
    reference_crossing_center=0.97135,
    reference_lower_bound=0.9711,
    initial_subdivisions=2,
    max_depth=1,
    min_width=1e-3,
    max_q=8,
    keep_last=1,
    q_min=2,
)
report['status']
```
