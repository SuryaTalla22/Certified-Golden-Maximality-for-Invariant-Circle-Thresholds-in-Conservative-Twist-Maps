```python
from pathlib import Path
import sys
PROJECT_ROOT = None
for cand in [Path.cwd(), Path.cwd().parent, *Path.cwd().parents]:
    if (cand / 'kam_theorem_suite').exists():
        PROJECT_ROOT = cand
        break
if PROJECT_ROOT is None:
    raise RuntimeError('Could not locate project root')
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
print('PROJECT_ROOT =', PROJECT_ROOT)
```

```python
from kam_theorem_suite import ArithmeticClassSpec
from kam_theorem_suite.proof_driver import (
    build_adaptive_class_campaign_report,
    build_seeded_class_campaign_report,
    build_seeded_multi_class_campaign_report,
)
from kam_theorem_suite.seed_transfer import build_seed_profile_from_orbit, build_seed_transfer_report
from kam_theorem_suite.standard_map import HarmonicFamily, solve_periodic_orbit
```

```python
family = HarmonicFamily()
src = solve_periodic_orbit(5, 8, 0.9711, family)
profile = build_seed_profile_from_orbit(src['x'], p=5, q=8, K=0.9711, family=family, label='5/8')
transfer = build_seed_transfer_report(profile, target_p=8, target_q=13, target_K=0.9712, family=family).to_dict()
{
    'source_label': transfer['source_label'],
    'target_label': transfer['target_label'],
    'selected_method': transfer['selected_method'],
    'selected_residual_inf': transfer['selected_residual_inf'],
    'candidate_methods': [row['method'] for row in transfer['candidate_scores']],
}
```

```python
silver = ArithmeticClassSpec(preperiod=(), period=(2,), label='silver')
adaptive = build_adaptive_class_campaign_report(
    silver,
    reference_lower_bound=0.97122,
    reference_crossing_center=0.97130,
    max_q=20,
    keep_last=2,
    q_min=2,
    auto_localize_crossing=True,
    min_width=1e-3,
    max_attempts_per_approximant=2,
)
seeded = build_seeded_class_campaign_report(
    silver,
    reference_lower_bound=0.97122,
    reference_crossing_center=0.97130,
    max_q=20,
    keep_last=2,
    q_min=2,
    auto_localize_crossing=True,
    min_width=1e-3,
    max_attempts_per_approximant=2,
)
{
    'adaptive_entries': len(adaptive['entries']),
    'seeded_entries': len(seeded['entries']),
    'seed_reuse_count': seeded['seed_reuse_count'],
    'seed_bank_size': len(seeded['seed_bank']),
    'seeded_certified_crossing_count': seeded['certified_crossing_count'],
}
```
