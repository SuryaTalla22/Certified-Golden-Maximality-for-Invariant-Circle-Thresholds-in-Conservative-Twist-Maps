# Golden supercritical obstruction and two-sided bridge

This notebook exercises the new golden-specific upper-side obstruction module and then pairs it with the golden a-posteriori lower-side continuation object.

```python
from pathlib import Path
import sys, json
_here = Path.cwd().resolve()
for cand in (_here, *_here.parents):
    if (cand / 'kam_theorem_suite').exists() or (cand / 'pyproject.toml').exists():
        if str(cand) not in sys.path:
            sys.path.insert(0, str(cand))
        root = str(cand)
        break
else:
    raise RuntimeError(f'Could not find project root from {_here}')
print(root)
from kam_theorem_suite import build_golden_supercritical_obstruction_report, build_golden_two_sided_bridge_report
```

```python
obstruction = build_golden_supercritical_obstruction_report(
    n_terms=5, keep_last=2, min_q=5, initial_subdivisions=1, max_depth=0,
)
print(json.dumps({
    'theorem_status': obstruction['theorem_status'],
    'selected_upper_source': obstruction['selected_upper_source'],
    'successful_crossing_count': obstruction['successful_crossing_count'],
    'successful_band_count': obstruction['successful_band_count'],
    'selected_upper_lo': obstruction['selected_upper_lo'],
    'selected_upper_hi': obstruction['selected_upper_hi'],
}, indent=2))
```
