# Two-Sided Irrational Threshold Corridors

This notebook exercises the new two-sided irrational atlas layer: a multiresolution irrational existence atlas on the lower side, and a multi-approximant rational obstruction ladder on the upper side.

```python
from pathlib import Path
import sys, math
ROOT = Path.cwd().resolve()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
print('ROOT =', ROOT)
```

```python
from kam_theorem_suite.obstruction_atlas import ApproximantWindowSpec
from kam_theorem_suite.proof_driver import build_rational_upper_ladder_report, build_two_sided_irrational_threshold_atlas_report
from kam_theorem_suite.standard_map import HarmonicFamily

GOLDEN = (math.sqrt(5.0) - 1.0) / 2.0
fam = HarmonicFamily()
specs = [
    ApproximantWindowSpec(3, 5, 0.9738, 0.9742, 1.04, 1.06, label='gold-3/5'),
    ApproximantWindowSpec(5, 8, 0.96, 0.98, 1.04, 1.06, label='gold-5/8'),
]
```

```python
ladder = build_rational_upper_ladder_report(
    rho=GOLDEN,
    specs=specs,
    family=fam,
    initial_subdivisions=2,
    max_depth=2,
)
{
    'ladder_quality': ladder['ladder_quality'],
    'successful_crossing_count': ladder['successful_crossing_count'],
    'successful_band_count': ladder['successful_band_count'],
    'best_crossing_interval': (ladder['best_crossing_lower_bound'], ladder['best_crossing_upper_bound']),
    'earliest_band_lo': ladder['earliest_band_lo'],
}
```

```python
atlas = build_two_sided_irrational_threshold_atlas_report(
    rho=GOLDEN,
    K_values=[0.5, 0.6, 0.7],
    specs=specs,
    family=fam,
    N_values=(32, 48, 64),
    quality_floor='weak',
    initial_subdivisions=2,
    max_depth=2,
)
{
    'status': atlas['relation']['status'],
    'lower_bound_K': atlas['relation']['lower_bound_K'],
    'upper_object': (atlas['relation']['upper_object_lo'], atlas['relation']['upper_object_hi']),
    'gap_lower_to_upper': atlas['relation']['gap_lower_to_upper'],
    'gap_lower_to_band': atlas['relation']['gap_lower_to_band'],
    'ladder_quality': atlas['relation']['ladder_quality'],
}
```

```python
# Show the upper-side entries compactly
[(e['label'], e['crossing_root_window_lo'], e['crossing_root_window_hi'], e['hyperbolic_band_lo'], e['status']) for e in atlas['upper_side']['approximants']]
```
