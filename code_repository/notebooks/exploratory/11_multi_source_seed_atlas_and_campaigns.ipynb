# Multi-Source Seed Atlas and Continuation Campaigns

Exercise the new atlas layer that blends several nearby successful approximants and compare it against the earlier single-source seeded workflow.


```python
from pathlib import Path
import sys
PROJECT_ROOT = None
for cand in [Path.cwd(), Path.cwd().parent, *Path.cwd().parents]:
    if (cand / 'kam_theorem_suite').exists():
        PROJECT_ROOT = cand
        break
if PROJECT_ROOT is None:
    raise RuntimeError('Could not locate project root')
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
print('PROJECT_ROOT =', PROJECT_ROOT)
```

```python
from kam_theorem_suite import ArithmeticClassSpec
from kam_theorem_suite.proof_driver import (
    build_multi_source_seed_atlas_report_driver,
    build_multi_source_class_campaign_report,
    build_multi_source_campaign_comparison_report,
    build_multi_source_multi_class_campaign_report,
)
from kam_theorem_suite.seed_transfer import build_seed_profile_from_orbit
from kam_theorem_suite.seeded_campaigns import SeedBankEntry
from kam_theorem_suite.standard_map import HarmonicFamily, solve_periodic_orbit
```

```python
family = HarmonicFamily()
seed_bank = []
for (p, q, K) in [(3, 5, 0.9710), (5, 8, 0.9711), (8, 13, 0.9712)]:
    sol = solve_periodic_orbit(p, q, K, family)
    profile = build_seed_profile_from_orbit(sol['x'], p=p, q=q, K=K, family=family, label=f'{p}/{q}')
    seed_bank.append(SeedBankEntry(label=f'{p}/{q}', p=p, q=q, rho=p / q, K_center=K, source_method='nb', seed_profile=profile.to_dict()))
atlas = build_multi_source_seed_atlas_report_driver(seed_bank, target_p=13, target_q=21, target_K=0.97125, family=family)
{
    'selected_method': atlas['selected_method'],
    'selected_source_labels': atlas['selected_source_labels'],
    'selected_residual_inf': atlas['selected_residual_inf'],
    'n_candidates': len(atlas['atlas_candidates']),
}
```

```python
silver = ArithmeticClassSpec(preperiod=(), period=(2,), label='silver')
atlas_campaign = build_multi_source_class_campaign_report(
    silver,
    reference_crossing_center=0.97130,
    reference_lower_bound=0.97122,
    max_q=20,
    keep_last=2,
    q_min=2,
    auto_localize_crossing=True,
    min_width=1e-3,
    max_attempts_per_approximant=2,
    max_sources=3,
)
{
    'atlas_use_count': atlas_campaign['atlas_use_count'],
    'atlas_blend_use_count': atlas_campaign['atlas_blend_use_count'],
    'certified_crossing_count': atlas_campaign['certified_crossing_count'],
    'status': atlas_campaign['status'],
}
```
